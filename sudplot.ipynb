```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("company_data.csv")
df["Order Date"]=pd.to_datetime(df['Order Date'] , format= "mixed")
df['Month'] = df['Order Date'].dt.month
df['Month'] =df['Order Date'].dt.strftime('%B')
df.to_csv("company_data.csv" ,index=False)
```

```python
fig,axs =plt.subplots(4,1,figsize =(8,10))
df=df.head(10)
# x=df.loc[0,20,'Order Date'] 

# y=df.loc[0,20,'Profit']

# a=df.loc[0,20,'Sales']

# b=df.loc[0,20,'Order Qty']

axs[0].plot(df['Month'],df['Sales'],'*:g',markerfacecolor='black')

axs[0].plot(title='Month and Profit',yladel='MWK')

axs[0].grid(True)

axs[1].plot(df['Month'],df['Profit'],'*:r',markerfacecolor='pink')

axs[1].plot(title='Month and profit',yladel='MWK')

axs[1].grid(True)

axs[2].plot(df['Month'],df['Order Qty'])
axs[2].plot(title='Month and Order',yladel='MWK')

axs[2].grid(True)

sorted_count=df.groupby('Country')['Order Qty'].sum().sort_values(ascending=False)
axs[3].bar(sorted_count.index,sorted_count.values)
axs[3].tick_params(rotation=45)
axs[3].grid(True)
# axs[3].plot(df.loc[0,20"mouth"])
plt.show()
```
