```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
def temperature_date():
    df=pd.read_csv("temperature_data.csv")
    print(df.head(10))
    df["Date Time"] = pd.to_datetime(df["Date Time"])
    x=df.loc[0:10,"Date Time"]
    y=df.loc[0:10,"T (degC)"]
    plt.plot(x,y,marker='*')
    plt.xticks(rotation=45)
    plt.title("A GRAPH OF TEMPERATURE OVER TIME")
    plt.xlabel('DATE TIME')
    plt.legend('Line') 
    plt.ylabel('TEMPERATURE')
    plt.show()
temperature_date()    
#ax.plot(x,y)
```

```python
#3
def exam_score():
    df=pd.read_csv("Student Exam Scores.csv")
    x=df[["MathScore","ReadingScore","WritingScore"]]
    plt.hist(x,bins=10)
    plt.title("A GRAPH OF EXAM SCORES AND FREQUENCY")
    plt.xlabel("Exam Scores")
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.legend(['bars'])
    plt.show
exam_score()    
```

```python
def Movies():
    df=pd.read_csv("Movies.csv")
    x=df.loc[0:10,"vote_count"]
    y=df.loc[0:10,"genres"]
    plt.pie(x,labels=y,autopct='%1.1f%%')
    # note("%1f%%") this is for only th e whole number
    plt.title("Movies")
    plt.xticks(rotation=45)
    plt.explode=[0.1,0,0,0,0] 
    # explode is used for removing the one with the biggest %
    plt.show
Movies()   
```

```python
def bmi():
    df=pd.read_csv("bmi.csv")
    x=df["Height"]
    y=df["Weight"]
    colour="green"
    plt.scatter(x,y,s=20,color=colour)
    plt.title("BMI")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show

bmi()
```
